```python
import requests

url = "https://fakestoreapi.com/products"

try:
    response = requests.get(url, timeout=10)
    response.raise_for_status()
    data = response.json()
    print(data[:2])
except requests.exceptions.RequestException as e:
    print("Error:", e)
```

```python
# Convert JSON to DataFrame
df = pd.json_normalize(data)

df.head()
```

```python
#rename column 
df = df[['id', 'title', 'price', 'category', 'rating.rate', 'rating.count']]

df.columns = ['ProductID', 'ProductName', 'Price', 'Category', 'Rating', 'RatingCount']

df.head()
```

```python
#export to csv
df.to_csv(r"E:\E-Commerce Sales Analysis\products.csv", index=False)
```
